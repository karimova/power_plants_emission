## Reading Data from .csv file

```python
# Read the mouse data and the study results
epp_data= pd.read_csv("data/database_epp_5years_selected_CSV_with_dots.csv", na_values = ".")
epp_data.head(10)
```

```python
new_df = epp_data.replace(np.nan, 0)
new_df.head(20)
```

```python
# Read the mouse data and the study results
epp_coords_data= pd.read_csv("data/december_generator2020_lan_long_CSV.csv")
epp_coords_data.head(20)
```

```python
#Removing dublicates from long-lan
no_dublicates = epp_coords_data.drop_duplicates(subset=['Plant Name'])
no_dublicates.head()
```

```python
merged_df = pd.merge(new_df, no_dublicates, on="Plant Id", how="inner")
merged_df.head()

merged_df = merged_df.rename(columns={"AER\nFuel Type Code": "AER_Fuel_Type_Code"})
```

```python
merged_df
```

```python
epp_data["AER\nFuel Type Code"].value_counts()
```

```python
merged_df['Plant Type'] = pd.np.where(merged_df.AER_Fuel_Type_Code.str.contains("NG"), "Natural Gas",
                        pd.np.where(merged_df.AER_Fuel_Type_Code.str.contains("SUN"), "Solar",
                        pd.np.where(merged_df.AER_Fuel_Type_Code.str.contains("GEO"), "Geothermal",
                        pd.np.where(merged_df.AER_Fuel_Type_Code.str.contains("WWW"), "Wood",
                        pd.np.where((merged_df.AER_Fuel_Type_Code.str.contains("COL")|merged_df.AER_Fuel_Type_Code.str.contains("WOC")), "Coal",
                        pd.np.where((merged_df.AER_Fuel_Type_Code.str.contains("DFO")|merged_df.AER_Fuel_Type_Code.str.contains("WOO")), "Petroleum",
                        pd.np.where((merged_df.AER_Fuel_Type_Code.str.contains("PC")|merged_df.AER_Fuel_Type_Code.str.contains("RFO")), "Petroleum",
                        pd.np.where(merged_df.AER_Fuel_Type_Code.str.contains("OOG"), "Natural Gas",
                        pd.np.where((merged_df.AER_Fuel_Type_Code.str.contains("HYC")|merged_df.AER_Fuel_Type_Code.str.contains("HPS")), "Hydroelectric",
                        pd.np.where(merged_df.AER_Fuel_Type_Code.str.contains("NUC"), "Nuclear",
                        pd.np.where(merged_df.AER_Fuel_Type_Code.str.contains("WND"), "Wind","Other")))))))))))
```

```python
merged_df.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
merged_df.to_csv("us_plants_coords.csv", index=False, header=True)
```
